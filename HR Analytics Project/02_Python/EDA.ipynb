```python
##EDA
##Employee Overview

import pandas as pd

df = pd.read_csv(r"E:\HR_Analytics_Project\01_Dataset\employee_attrition_dataset_10000.csv")

total_employees = len(df)
employees_left = df[df['Attrition'] == 'Yes'].shape[0]
employees_stayed = df[df['Attrition'] == 'No'].shape[0]

attrition_rate = (employees_left / total_employees) * 100

print("Total Employees :", total_employees)
print("Employees Left :", employees_left)
print("Employees Stayed :", employees_stayed)
print(f"Attrition Rate : {attrition_rate:.2f}%")
```

```python
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.countplot(data=df, x='Department', hue='Attrition')

plt.title("Department-wise Attrition")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=30)
plt.show()
```

Observation:
    The "Marketing" department has the highest employee attrition, indicating possible workload or job satisfaction issues.
    The "IT" department has the lowest attrition, suggesting better employee retention.
    HR should investigate the reasons behind higher attrition in affected departments.

```python
##Salary Band Analysis
df['Salary_Band'] = pd.cut(
    df['Monthly_Income'],
    bins=[0,5000,10000,20000],
    labels=['Low','Medium','High']
)
```

```python
##Count Plot

plt.figure(figsize=(8,5))

sns.countplot(data=df,x='Salary_Band',hue='Attrition')

plt.title("Salary Band vs Attrition")
plt.xlabel("Salary Band")
plt.ylabel("Employee Count")
plt.show()
```

Observation:
    Employees in the Low Salary band show higher attrition compared to Medium and High salary bands.
    Higher salaries appear to improve employee retention.
    Reviewing compensation for lower-paid employees may help reduce attrition.

```python
## Promotion Analysis

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Attrition',
    y='Years_Since_Last_Promotion'
)

plt.title("Years Since Last Promotion vs Attrition")
plt.show()
```

Observation:
    Employees who have not received promotions for a longer period tend to leave the organization more frequently.
    Regular promotions and career growth opportunities can improve employee retention.
    Performance-based promotion policies may help reduce attrition.

```python
## Overtime Analysis

plt.figure(figsize=(7,5))

sns.countplot(data=df,x='Overtime',hue='Attrition')

plt.title("Overtime vs Attrition")
plt.show()
```

Observation:
    Employees working overtime have a higher attrition rate than those who do not.
    Excessive overtime may contribute to employee stress and burnout.
    Balancing workloads could improve employee satisfaction and retention.

```python
## Job Satisfaction Analysis

plt.figure(figsize=(8,5))

sns.countplot(data=df,x='Job_Satisfaction',hue='Attrition')

plt.title("Job Satisfaction vs Attrition")
plt.show()
```

Observation:
    Employees with lower job satisfaction are more likely to leave the organization.
    Higher satisfaction levels are associated with lower attrition.
    Improving workplace culture and employee engagement may increase retention.

```python
## Work-life Balance Analysis

plt.figure(figsize=(8,5))

sns.countplot(data=df,x='Work_Life_Balance',hue='Attrition')

plt.title("Work Life Balance vs Attrition")
plt.show()
```

Observation:
    Employees reporting poor work-life balance show higher attrition.
    Better work-life balance is associated with improved employee retention.
    Flexible work arrangements may help reduce employee turnover.

```python
## Age Analysis


plt.figure(figsize=(8,5))

sns.boxplot(data=df,x='Attrition',y='Age')

plt.title("Age vs Attrition")
plt.show()
```

Observation
    The median age of employees who stayed (No) and those who left (Yes) is approximately 40 years, indicating that attrition occurs across similar age groups.
    The age distribution for both groups is very similar, with most employees falling between 30 and 50 years of age.
    Since there is no significant difference in the age distributions, age alone does not appear to be a major factor influencing employee attrition. Other factors such as salary, overtime, promotions, and job satisfaction are likely to have a greater impact on employee turnover.

```python
## Gender Analysis


plt.figure(figsize=(7,5))

sns.countplot(data=df,x='Gender',hue='Attrition')

plt.title("Gender vs Attrition")
plt.show()
```

```python
## Job Role Analysis

plt.figure(figsize=(12,5))

sns.countplot(data=df,x='Job_Role',hue='Attrition')

plt.xticks(rotation=45)

plt.title("Job Role vs Attrition")
plt.show()
```

Observation:
    The "Analyst" job role has the highest attrition.
    Some job roles experience significantly better retention than others.
    HR should investigate role-specific challenges to reduce turnover.

```python
## Correlation Heatmap

plt.figure(figsize=(14,10))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

Observation:
    Most numerical features show weak to moderate correlations, indicating that attrition is influenced by multiple factors.
    Variables such as income, experience, and promotions may have stronger relationships with employee retention.
    Machine learning can better capture these combined effects than simple correlation analysis.
